```python

def classes(api_key,class_name):
    all_records = []

    for i in range(1, 26):
        url = "https://api.harvardartmuseums.org/object"
        params = {
            "apikey": api_key,
            "size": 100,
            "page": i,
            "classification": class_name
        }
        response = requests.get(url, params)

        data = response.json()
        records = data.get('records', [])
        all_records.extend(records)

        return all_records

records = classes(api_key, "Paintings")
records
```

```python

artifact_metadata = []    
for i in records:
       artifact_metadata.append(dict(id = i['id'],
       title = i['title'],
       culture = i['culture'],
       period = i['period'],
       century = i['century'],
       medium = i['medium'],
       dimensions = i['dimensions'],
       description = i['description'],
       department = i['department'],
       classification = i['classification'],
       aaccessionyear = i['accessionyear'],
       accessionmethod = i['accessionmethod']))

artifact_metadata


   
```

```python

artifact_media = []
for i in records:
       artifact_media.append(dict(objectid = i['objectid'],
       imagecount = i['imagecount'],
       mediacount = i['mediacount'],
       colorcount = i['colorcount'],
       rank_num = i['rank'],
       datebegin = i['datebegin'],
       dateend = i['dateend']))
artifact_media       
```

```python

artifact_colors = []

for i in records:
       if isinstance(i, dict) and 'objectid' in i:
        for j in i.get('colors', []): 
            if isinstance(j, dict): 
                artifact_colors.append({
                    'objectid': i['objectid'],
                    'color': j.get('color'),
                    'spectrum': j.get('spectrum'),
                    'hue': j.get('hue'),
                    'percent': j.get('percent'),
                    'css3': j.get('css3')
                })
artifact_colors
```
